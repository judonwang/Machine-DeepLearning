```python
import numpy as np  # Import NumPy for RMSE calculation
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from torch.utils.data import DataLoader, Dataset

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Load your final merged dataset
data = final_data  # Adjust to your final_data

# Preprocess categorical features (without encoding 'anime_id' and 'user_id')
label_encoders = {}
for column in ['Genres', 'Type', 'Source', 'Status']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column].astype(str))
    label_encoders[column] = le

# Normalize numerical features
scaler = StandardScaler()
num_cols = ['Score', 'Rank', 'Popularity', 'Favorites', 'Days Watched', 'Mean Score', 'Episodes Watched']
data[num_cols] = scaler.fit_transform(data[num_cols])

# Target variable
data['target'] = data['rating'].fillna(data['rating'].mean())

# Split data
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Custom Dataset Class with user_id and anime_id
class AnimeDataset(Dataset):
    def __init__(self, df):
        self.user_ids = torch.tensor(df['user_id'].values, dtype=torch.long)
        self.anime_ids = torch.tensor(df['anime_id'].values, dtype=torch.long)
        self.genres = torch.tensor(df['Genres'].values, dtype=torch.long)
        self.type = torch.tensor(df['Type'].values, dtype=torch.long)
        self.source = torch.tensor(df['Source'].values, dtype=torch.long)
        self.status = torch.tensor(df['Status'].values, dtype=torch.long)
        self.numerical_data = torch.tensor(df[num_cols].values, dtype=torch.float32)
        self.labels = torch.tensor(df['target'].values, dtype=torch.float32)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return (self.user_ids[idx], self.anime_ids[idx], self.genres[idx], self.type[idx],
                self.source[idx], self.status[idx], self.numerical_data[idx]), self.labels[idx]


train_dataset = AnimeDataset(train_data)
test_dataset = AnimeDataset(test_data)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)

class WideAndDeepModel(nn.Module):
    def __init__(self, num_users, num_anime, num_genres, num_type, num_source, num_status):
        super(WideAndDeepModel, self).__init__()

        # Embeddings for categorical features
        embedding_dim = 16  # Increased for better representation
        self.user_embedding = nn.Embedding(num_users, embedding_dim)
        self.anime_embedding = nn.Embedding(num_anime, embedding_dim)
        self.genre_embedding = nn.Embedding(num_genres, embedding_dim)
        self.type_embedding = nn.Embedding(num_type, embedding_dim)
        self.source_embedding = nn.Embedding(num_source, embedding_dim)
        self.status_embedding = nn.Embedding(num_status, embedding_dim)

        # Deep part (MLP) layers
        self.deep = nn.Sequential(
            nn.Linear(embedding_dim * 6 + len(num_cols), 128),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.Dropout(0.3),
            nn.Linear(64, 1)
        )

        # Wide part
        self.wide = nn.Linear(len(num_cols), 1)

    def forward(self, x):
        user_id, anime_id, genres, anime_type, source, status, num_data = x

        # Embedding lookups
        user_emb = self.user_embedding(user_id)
        anime_emb = self.anime_embedding(anime_id)
        genre_emb = self.genre_embedding(genres)
        type_emb = self.type_embedding(anime_type)
        source_emb = self.source_embedding(source)
        status_emb = self.status_embedding(status)

        # Concatenate embeddings for deep component
        deep_input = torch.cat([user_emb, anime_emb, genre_emb, type_emb, source_emb, status_emb, num_data], dim=1)
        deep_out = self.deep(deep_input)

        # Wide component
        wide_out = self.wide(num_data)

        # Combine wide and deep outputs
        out = deep_out + wide_out
        return out.squeeze()

# Initialize model with the max user_id and anime_id for embedding dimensions
model = WideAndDeepModel(
    num_users=data['user_id'].max() + 1,       # Set embedding size to max user_id + 1
    num_anime=data['anime_id'].max() + 1,      # Set embedding size to max anime_id + 1
    num_genres=len(label_encoders['Genres'].classes_),
    num_type=len(label_encoders['Type'].classes_),
    num_source=len(label_encoders['Source'].classes_),
    num_status=len(label_encoders['Status'].classes_)
).to(device)  # Move model to device

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005)

# Training and Evaluation Functions
def train(model, train_loader, criterion, optimizer):
    model.train()
    running_loss = 0.0
    total_predictions = 0

    for inputs, labels in train_loader:
        # Move inputs and labels to device
        inputs = tuple(tensor.to(device) for tensor in inputs)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * labels.size(0)  # Accumulate total loss for RMSE calculation
        total_predictions += labels.size(0)

    avg_loss = running_loss / total_predictions
    rmse = np.sqrt(avg_loss)  # Calculate RMSE
    return avg_loss, rmse

def evaluate(model, test_loader, criterion):
    model.eval()
    running_loss = 0.0
    total_predictions = 0

    with torch.no_grad():
        for inputs, labels in test_loader:
            # Move inputs and labels to device
            inputs = tuple(tensor.to(device) for tensor in inputs)
            labels = labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * labels.size(0)  # Accumulate total loss for RMSE calculation
            total_predictions += labels.size(0)

    avg_loss = running_loss / total_predictions
    rmse = np.sqrt(avg_loss)  # Calculate RMSE
    return avg_loss, rmse

# Training Loop
epochs = 15
for epoch in range(epochs):
    train_loss, train_rmse = train(model, train_loader, criterion, optimizer)
    test_loss, test_rmse = evaluate(model, test_loader, criterion)
    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, Train RMSE: {train_rmse:.4f}, "
          f"Test Loss: {test_loss:.4f}, Test RMSE: {test_rmse:.4f}")
```

```python
# Personalized Recommendation Function
def recommend(user_id, model, all_anime_ids, top_k=10):
    # Convert user_id to tensor and repeat it to match all anime IDs
    user = torch.tensor([user_id] * len(all_anime_ids), dtype=torch.long).to(device)
    anime_ids = torch.tensor(all_anime_ids, dtype=torch.long).to(device)

    # Placeholder tensors for other categorical inputs
    genres = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    anime_type = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    source = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    status = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder

    # Placeholder for numerical features
    num_data = torch.zeros((len(all_anime_ids), len(num_cols)), dtype=torch.float32).to(device)

    # Bundle inputs as expected by the model's forward method
    inputs = (user, anime_ids, genres, anime_type, source, status, num_data)

    # Switch model to evaluation mode
    model.eval()
    with torch.no_grad():
        # Get predictions for all anime IDs for the given user
        scores = model(inputs)

    # Select the top-k anime based on predicted scores
    _, top_indices = torch.topk(scores, top_k)
    recommended_anime_ids = anime_ids[top_indices].tolist()

    return recommended_anime_ids

# Example usage: Get top 10 recommendations for a specific user (e.g., user_id = 1)
sample_user_id = 1
all_anime_ids = list(range(data['anime_id'].nunique()))  # Assuming anime IDs are sequential
recommendations = recommend(sample_user_id, model, all_anime_ids, top_k=10)
print("Top 10 recommended anime IDs for user", sample_user_id, ":", recommendations)
```

The recommend function provided does not actually check if user_id = 10 is present in the dataset. Instead, it simply assumes that any specified user_id exists and generates recommendations for it based on the model's learned weights. This behavior is common in recommendation systems since models can make predictions for users regardless of their presence in the training data.

If user_id = 10 is not in the training data, the model will use the embeddings and patterns it learned from other users to generate recommendations for user_id = 10. This is known as generalization in machine learning, where a model attempts to make meaningful predictions even for unseen inputs.

```python
# Step 1: Create a dictionary mapping anime_id to anime_name
anime_id_to_name = dict(zip(final_data['anime_id'], final_data['Name']))

# Updated Recommendation Function
def recommend(user_id, model, all_anime_ids, top_k=10):
    # Convert user_id to tensor and repeat it to match all anime IDs
    user = torch.tensor([user_id] * len(all_anime_ids), dtype=torch.long).to(device)
    anime_ids = torch.tensor(all_anime_ids, dtype=torch.long).to(device)

    # Placeholder tensors for other categorical inputs
    genres = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    anime_type = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    source = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder
    status = torch.zeros(len(all_anime_ids), dtype=torch.long).to(device)  # Placeholder

    # Placeholder for numerical features
    num_data = torch.zeros((len(all_anime_ids), len(num_cols)), dtype=torch.float32).to(device)

    # Bundle inputs as expected by the model's forward method
    inputs = (user, anime_ids, genres, anime_type, source, status, num_data)

    # Switch model to evaluation mode
    model.eval()
    with torch.no_grad():
        # Get predictions for all anime IDs for the given user
        scores = model(inputs)

    # Select the top-k anime based on predicted scores
    _, top_indices = torch.topk(scores, top_k)
    recommended_anime_ids = anime_ids[top_indices].tolist()

    # Map anime IDs to their names and print recommendations
    recommendations = [(anime_id, anime_id_to_name.get(anime_id, "Unknown")) for anime_id in recommended_anime_ids]
    return recommendations

# Example usage: Get top 10 recommendations for a specific user (e.g., user_id = 1)
sample_user_id = 10
all_anime_ids = list(range(final_data['anime_id'].nunique()))  # Assuming anime IDs are sequential
recommendations = recommend(sample_user_id, model, all_anime_ids, top_k=10)

# Print recommendations with names
print("Top 10 recommended anime for user", sample_user_id, ":")
for anime_id, Name in recommendations:
    print(f"Anime ID: {anime_id}, Anime Name: {Name}")
```
