```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('wordnet')
stop_word_group = stopwords.words('english')

lematizer = WordNetLemmatizer()

def preprocess_text(textblob):
    textblob = re.sub(r'\d', '', textblob, re.UNICODE)
    textblob = re.sub(r'[^\w\s]', '', textblob, re.UNICODE)
    textblob = textblob.lower()
    textblob = [lematizer.lemmatize(token) for token in textblob.split(" ")]
    textblob = [word for word in textblob if not word in stop_word_group] # removing any stopwords 
    textblob = " ".join(textblob)
    return textblob

# def preprocess_text(textblob):
#     try:
#         textblob = re.sub(r'\d', '', textblob, re.UNICODE)
#         textblob = re.sub(r'[^\w\s]', '', textblob, re.UNICODE)
#         textblob = textblob.lower()
#         textblob = [lematizer.lemmatize(token) for token in textblob.split(" ")]
#         textblob = [word for word in textblob if not word in stop_word_group] # removing any stopwords 
#         textblob = " ".join(textblob)
#     except:
#         print(textblob)
#         return textblob
#     return textblob


df3['reviewText'] = df3.reviewText.apply(lambda x: preprocess_text(x))
df3
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vvecTool = TfidfVectorizer(max_features=2000)
data_features = vvecTool.fit_transform(df3['reviewText'])
data_features = data_features.toarray()
print(data_features)
```

```python
y = df3['overall']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_features, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.ensemble import RandomForestClassifier

mod = RandomForestClassifier()
param_grid = {
    'n_estimators': [10, 20, 50, 100, 200],
    'max_depth': [None, 5, 10]
}
r_search = RandomizedSearchCV(estimator=mod, param_distributions=param_grid, n_iter=3, cv=2, verbose=2, random_state=42, n_jobs=-1)

r_search.fit(X_train, y_train)

best_params = r_search.best_params_
print("Best parameters:", best_params)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=10, random_state=42, max_depth=None)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_roc = model.predict_proba(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score


print("Precision:", precision_score(y_test, y_pred, average='macro'))

print("Recall:", recall_score(y_test, y_pred, average='macro'))

print("AUC: ", roc_auc_score(y_test, y_pred_roc, average = 'macro', multi_class='ovr'))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```
