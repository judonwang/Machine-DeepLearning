```python
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('wordnet')
stop_word_group = stopwords.words('english')

lematizer = WordNetLemmatizer()

def preprocess_text(textblob):
    textblob = re.sub(r'\d', '', textblob, re.UNICODE)
    textblob = re.sub(r'[^\w\s]', '', textblob, re.UNICODE)
    textblob = textblob.lower()
    textblob = [lematizer.lemmatize(token) for token in textblob.split(" ")]
    textblob = [word for word in textblob if not word in stop_word_group] # removing any stopwords 
    textblob = " ".join(textblob)
    return textblob

# def preprocess_text(textblob):
#     try:
#         textblob = re.sub(r'\d', '', textblob, re.UNICODE)
#         textblob = re.sub(r'[^\w\s]', '', textblob, re.UNICODE)
#         textblob = textblob.lower()
#         textblob = [lematizer.lemmatize(token) for token in textblob.split(" ")]
#         textblob = [word for word in textblob if not word in stop_word_group] # removing any stopwords 
#         textblob = " ".join(textblob)
#     except:
#         print(textblob)
#         return textblob
#     return textblob


df3['reviewTextSummary'] = df3.reviewTextSummary.apply(lambda x: preprocess_text(x))
df3
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vvecTool = TfidfVectorizer(max_features=2000)
data_features = vvecTool.fit_transform(df3['reviewTextSummary'])
data_features = data_features.toarray()
print(data_features)
```

```python
y = df3['overall']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_features, y, test_size=0.2, random_state=42)
```

```python
# downsampled data with similar pattern of positive to neutral to negative reviews
df_maj = df3[df3['overall'] == 2]
df_min_1 = df3[df3['overall'] == 1]
df_min_0 = df3[df3['overall'] == 0]

df_maj_down_2 = df_maj.sample(frac=0.21, random_state=42)

df_min1_up_2 = df_min_1.sample(n=len(df_maj_down_2) * 4 // 8, replace=True, random_state=42)
df_min0_up_2 = df_min_0.sample(n=len(df_maj_down_2) * 3 // 8, replace=True, random_state=42)

df_resample_2 = pd.concat([df_maj_down_2, df_min1_up_2, df_min0_up_2])

df_resample_2
```

```python
X_resample_2 = vvecTool.transform(df_resample_2['reviewTextSummary'])
y_resample_2 = df_resample_2['overall']
```

```python
from sklearn.model_selection import RandomizedSearchCV
# from scipy.stats import randint
# from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


#mod = XGBClassifier(objective='multi:softmax', num_class=3, tree_method='gpu_hist')

mod = XGBClassifier(objective='multi:softmax', num_class=3, tree_method='hist', device='cuda')
param_grid = {
    'max_depth': [10, 20, 30, 40, 50, 60, None],
    'n_estimators': [100, 200, 400, 800, 1000],
    'learning_rate': [0.01, 0.1, 0.2],
    'gamma': [0.1, 0.5, 1],    
}
r_search = RandomizedSearchCV(estimator=mod, param_distributions=param_grid, n_iter=14, cv=5, verbose=3, random_state=42, n_jobs=-1)

r_search.fit(X_resample_2, y_resample_2)

best_params = r_search.best_params_
print("Best parameters:", best_params)
```

```python
from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)
sample_weights
```

```python
from sklearn.metrics import roc_curve
import xgboost as xgb
from xgboost import XGBClassifier

model = XGBClassifier(**best_params, objective='multi:softmax', n_class=3, random_state=42, tree_method='hist', device='cuda')
#, ojective='multi:softprob', eval_metric='auc'


model.fit(X_train, y_train, sample_weight=sample_weights)
```

```python
y_pred = model.predict(X_test)
y_pred_roc = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("AUC: ", roc_auc_score(y_test, y_pred_roc, average = 'macro', multi_class='ovr'))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```
